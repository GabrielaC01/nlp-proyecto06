Cargamos las librer√≠as

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Generamo la secuencias binarias

```python
# Usamos una semilla fija para que las secuencias generadas sean siempre las mismas
np.random.seed(42)
sequences = [np.random.randint(0, 2, size=np.random.randint(4, 11)).tolist() for _ in range(100)]

for i, seq in enumerate(sequences[:5], 1):
    print(f"Secuencia {i}: {seq}")
```

Analizamos las longitudes de las secuencias

```python
lengths = [len(seq) for seq in sequences]
lengths_df = pd.DataFrame(lengths, columns=["longitud"])

print(lengths_df.describe())
```

```python
plt.hist(lengths, bins=7, color='green', edgecolor='black')
plt.title("Distribuci√≥n de la longitud de las secuencias binarias")
plt.xlabel("Longitud de la secuencia")
plt.ylabel("Frecuencia")
plt.show()
```
